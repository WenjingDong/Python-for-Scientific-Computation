a) From the pictures above, we can see
Increasing the order of the scheme always improves the solution.
But decreasing the time step does not always improve the solution for the result is very close to the exact solution when time step becomes small enough and the result changes little when the time step becomes even smaller.

b)Move the exact solution to make the exact solution coincides with the approximated solution at $ t_i $  , then the local truncation at $ t_i $ is the value of exact solution minus that of approximated solution at  $ t_{i+1} $.The global error is the value of exact solution at $ t_n $ minus that of the approximated solution

c)For different grid spcing, that is, different $ \Delta t $, we will be able to have different local truncation errors and global erros correspondly.
Assume we have  $ \Delta t_1 $, $ \Delta t_2 $, $ \Delta t_3 $, correspondly local truncation errors: $ y(\Delta t_1)$,  $ y(\Delta t_2)$, $ y(\Delta t_3)$, and global errors: $ g(\Delta t_1)$,  $ g(\Delta t_2)$, $ g(\Delta t_3)$.
And let y= local truncation error,g=global erro, both of which are the functions of  $ \Delta t $, then we have:
$$y=c(\Delta t)^p_1$$ and $$g=c(\Delta t)^p_2$$
That is, 
$$ln|y|=ln|c|+p_1ln\Delta t$$
and $$ln|y|=ln|c|+p_2ln\Delta t$$
Plot $(ln(\Delta t_1),ln|y(\Delta t_1)|)$, $(ln(\Delta t_2),ln|y(\Delta t_2)|)$, $(ln(\Delta t_2),ln|y(\Delta t_2)|)$, by linear fit we will get $p_1$, which is the slope's tangent of the linear fit. So we get the order of the local truncation error.
Similarly, we can get  $p_2$, the order of global error.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numlabs.lab2.lab2_functions import euler,beuler,leapfrog
import numpy as np


theFuncs={'euler':euler,'beuler':beuler,'leapfrog':leapfrog}

if __name__=="__main__":
    tend=20.
    Ta=20.
    To=30.
    theLambda=-8.
    funChoice='euler'
    npts=80.
    approxTime,approxTemp=theFuncs[funChoice](npts,tend,To,Ta,theLambda)
    exactTime=np.empty([npts,],float)
    exactTemp=np.empty_like(exactTime)
    for i in np.arange(0,npts):
       exactTime[i] = tend*i/npts
       exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
    plt.close('all')
    plt.figure(1)
    plt.clf()
    plt.plot(exactTime,exactTemp,'r+')
    plt.hold(True)
    plt.plot(approxTime,approxTemp,label='forward euler')
    theAx=plt.gca()
    theAx.set_xlim([0,20])
    theAx.set_ylim([15,30])
    theAx.set_title('stability-forward euler with deltat=0.25')
    plt.show()
```
